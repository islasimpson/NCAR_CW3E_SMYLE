```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from math import nan
import pandas as pd
from dateutil.relativedelta import relativedelta
```

```python
pathout="/glade/campaign/cgd/cas/islas/python_savs/NCAR_CW3E_SMYLE/DATA_SORT/pacific_zmU/"
basepath="/glade/campaign/cgd/cas/islas/python_savs/NCAR_CW3E_SMYLE/DATA_SORT/U_1000_10/"
ystart=1970
yend=2020
```

### Output L83

```python
datl83=[]
for iyear in np.arange(ystart,yend+1,1):
    print(iyear)
    dat = xr.open_dataset(basepath+'L83/U_1000_10_L83_init11_'+str(iyear)+'.nc').rename({'plev':'pre'})
    dat = dat.sel(lon=slice(130,230)).mean('lon')
    datl83.append(dat)
datl83 = xr.concat(datl83, dim='init_year')
datl83['init_year'] = np.arange(ystart,yend+1,1)
datl83.to_netcdf(pathout+'U_pacific_L83_init11.nc')
```

### Output L32

```python
datl32=[]
for iyear in np.arange(ystart,yend+1,1):
    print(iyear)
    dat = xr.open_dataset(basepath+'L32/U_1000_10_L32_init11_'+str(iyear)+'.nc').rename({'plev':'pre'})
    dat = dat.sel(lon=slice(130,230)).mean('lon')
    datl32.append(dat)
datl32 = xr.concat(datl32, dim='init_year')
datl32['init_year'] = np.arange(ystart,yend+1,1)
datl32.to_netcdf(pathout+'U_pacific_L32_init11.nc')
```

### Output ERA5

```python
endmon=4
dat = xr.open_mfdataset("/glade/campaign/cgd/cas/observations/ERA5/mon_f09/ua/*.nc")
alldat=[]
for iyear in np.arange(ystart,yend+1,1):
    datuse = dat.sel(time=slice(str(iyear)+'-11-01', str(iyear+1)+'-'+str(endmon).zfill(2)+'-30'))

    if (iyear == ystart):
        datestart = pd.to_datetime("1970-"+str(datuse.isel(time=0).time.dt.month.values).zfill(2)+"-"+str(datuse.isel(time=0).time.dt.day.values).zfill(2),
                                   format="%Y-%m-%d")
        time = [ datestart + relativedelta(months=int(i)) for i in np.arange(0,datuse.time.size,1) ]

    datuse['time'] = time
    datuse = datuse.sel(lon=slice(130,230)).mean('lon').load()
    alldat.append(datuse)
alldat = xr.concat(alldat, dim='init_year')
alldat['init_year'] = np.arange(ystart,yend+1,1)
alldat.to_netcdf(pathout+'U_pacific_ERA5_init11.nc')
```
